#Data cleaning

```python
loc['country']=loc['country'].fillna("unknown")
```

```python
loc['director']=loc['director'].fillna("unknown")
```

```python
loc=loc.dropna(subset=['date_added','rating'])
```

```python
loc.isnull().sum()
```

```python
loc['date_added']=pd.to_datetime(loc['date_added'])
```

```python
loc
```

```python
loc['year_adedd']=loc['date_added'].dt.year
```

#visualization

```python
sns.countplot(data=loc,x='type',hue='type',palette='deep')
plt.title("Movis vs TV Shows")
plt.ylabel("")
plt.show
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=loc,x='rating',hue='rating',palette='deep')
plt.title("Who can whatch")
plt.ylabel("")
plt.show
```

```python
loc.head(10)
```

```python
top_countries=loc['country'].value_counts().head(10)
```

```python
top_countries
```

```python
sns.barplot(x=top_countries.values,y=top_countries.index,hue=top_countries)
plt.ylabel("")
plt.show()
```

```python
top_director=loc['director'].value_counts().head(5)
```

```python
top_director
```

```python
sns.barplot(x=top_director.values,y=top_director.index,hue=top_director)
plt.ylabel("")
plt.show()
```

```python
sns.histplot(data=loc,x=loc['release_year'],palette='warm')
plt.title("relase over years")
plt.ylabel("")
plt.show()
```

```python
sns.histplot(data=loc,x=loc['year_adedd'],bins=0)
plt.title("year_adedd")
plt.ylabel("")
plt.show()
```
